```python
df["month"]=pd.to_datetime(df["Order Date"])
```

```python
df.info()
```

```python
df["month_year"]=df["month"].apply (lambda x: x.strftime("%y-%m"))
```

```python
df["month_year"]
```

```python
df.head()
```

```python
df_trend=df.groupby("month_year").sum()["Sales"].reset_index()
```

```python
df_trend
```

```python
df["month"].max()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(15,6))
plt.plot(df_trend["month_year"],df_trend["Sales"],color="red")
plt.xticks(rotation=90)
plt.yticks(rotation=45)
plt.xlabel("2015-01-02 To 2018-12-30",fontsize=15)
plt.ylabel("Sales",fontsize=15)
plt.show()
```
